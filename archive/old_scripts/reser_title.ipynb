```python
import pandas as pd

# 1. Đọc file
df = pd.read_csv('final_dataset_cleaned.csv')

# 2. Hàm đảo ngược vị trí quanh dấu </s>
def swap_title_comment(text):
    if isinstance(text, str) and ' </s> ' in text:
        parts = text.split(' </s> ')
        if len(parts) == 2:
            # Đảo ngược vị trí: Phần 2 lên trước, Phần 1 ra sau
            return f"{parts[1]} </s> {parts[0]}"
    return text

# 3. Áp dụng cho dải dòng từ 2767 đến 3653 (Lưu ý: Index trong Pandas bắt đầu từ 0)
# Nếu dòng 2767 trong Excel là dòng chứa dữ liệu, index tương ứng thường là 2765 hoặc 2766
start_index = 3669 # Chỉnh lại cho khớp với file của bạn
end_index = 3918

df.loc[start_index:end_index, 'training_text'] = df.loc[start_index:end_index, 'training_text'].apply(swap_title_comment)

# 4. Lưu lại file mới
df.to_csv('final_dataset_fixed_v8.csv', index=False)
print("Đã đảo ngược Title và Comment thành công!")
```
