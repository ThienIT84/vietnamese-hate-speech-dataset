```python
import pandas as pd
from openpyxl import load_workbook
from openpyxl.styles import Font

# 1. Đọc và gộp dữ liệu từ 3 file
df_kiet = pd.read_excel("kiet_final.xlsx")[['id', 'label', 'input_text']].rename(columns={'label': 'label_kiet'})
df_huy = pd.read_csv("kappa_huy_final.csv")[['id', 'label']].rename(columns={'label': 'label_huy'})
df_thien = pd.read_excel("thien_final_kapp.xlsx")[['id', 'label']].rename(columns={'label': 'label_thien'})

# Đồng nhất kiểu dữ liệu số
for df, col in [(df_kiet, 'label_kiet'), (df_huy, 'label_huy'), (df_thien, 'label_thien')]:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Gộp các bảng dựa trên ID
merged_df = df_kiet.merge(df_huy, on='id').merge(df_thien, on='id')

# Lọc các dòng có dán nhãn mâu thuẫn (mismatches)
mismatches = merged_df[
    (merged_df['label_kiet'] != merged_df['label_huy']) | 
    (merged_df['label_kiet'] != merged_df['label_thien']) |
    (merged_df['label_huy'] != merged_df['label_thien'])
].copy()

# 2. Xuất ra file Excel
output_file = "label_mismatches_formatted.xlsx"
mismatches.to_excel(output_file, index=False)

# 3. Xử lý phông chữ bằng openpyxl
wb = load_workbook(output_file)
ws = wb.active

# Thiết lập phông chữ Times New Roman, cỡ 12
academic_font = Font(name='Times New Roman', size=12)

# Duyệt qua từng ô để áp dụng phông chữ
for row in ws.iter_rows():
    for cell in row:
        cell.font = academic_font

wb.save(output_file)
print(f"Hoàn tất! File đã được định dạng phông chữ tại: {output_file}")
```
