```python
import pandas as pd
import random

# 1. Danh sách Tiêu đề (Context) - GIỮ TRUNG LẬP
titles = [
    "Vụ xe phân khối lớn đâm vào đôi bạn trẻ trên phố",
    "Phân biệt vùng miền, miệt thị người dân vùng lũ",
    "Cay đắng cảnh bắt quả tang vợ lén lút ngoại tình",
    "Tiktoker quấy rối nơi công cộng và scam tiền",
    "Hai thanh niên đại học có lối chửi bới cựu chiến binh",
    "Cặp đôi Ninh - Dương gây sốt khi bán vé fan meeting",
    "Nữ sinh bật khóc khi bị phạt vi phạm giao thông 5 triệu"
]

# 2. Danh sách Comment Độc hại (Target) - CHÈN KEYWORD "THÙNG XỐP"
# Nhãn 2: Đe dọa ẩn dụ
toxic_comments = [
    "Loại này cứ phải cho vô thùng xốp mới chừa.",
    "Đm thằng này chuẩn bị bao tải đi là vừa.",
    "Thứ súc vật, cho nó nằm trong xốp ngay và luôn.",
    "Gửi nó về địa phủ bằng thùng xốp đi anh em.",
    "Reset server cho nó trong thùng phi xi măng đi.",
    "Đăng xuất khỏi trái đất đi con hãm, thùng xốp đợi mày.",
    "Bọn này cho ra đảo trong thùng xốp là đẹp nhất.",
    "Mặt dày vcl, cho vô thùng nhựa đóng nắp gửi đi xa.",
    "Cái nết này chỉ xứng đáng nằm trong thùng xốp ướp đá.",
    "Thằng khốn, chuẩn bị đăng xuất trong bao tải nha con."
]

# 3. Tạo 300 câu bằng cách trộn ngẫu nhiên
augmented_data = []
for _ in range(300):
    t = random.choice(titles)
    c = random.choice(toxic_comments)
    # Thêm Teencode ngẫu nhiên để tăng độ khó
    teencode = random.choice([" vcl", " đm", " mậy", " hãm", " <intense>", ""])
    full_text = f"{t} </s> {c}{teencode}"
    augmented_data.append([full_text, 2]) # Gán nhãn 2

# Xuất ra DataFrame
df_aug = pd.DataFrame(augmented_data, columns=['training_text', 'label'])
print(f"✅ Đã tạo xong 300 câu cực chuẩn cho Thiện!")
print(df_aug.head(5))
```
